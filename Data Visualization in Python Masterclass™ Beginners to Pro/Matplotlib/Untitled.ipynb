```python
# Pie chart, where the slices will be ordered and plotted counter-clockwise:
labels = 'Frogs', 'Hogs', 'Dogs', 'Logs'
sizes = [15, 30, 45, 15]
explode = (0, 0.1, 0, 0)  # only "explode" the 2nd slice (i.e. 'Hogs')

fig1, ax1 = plt.subplots()
ax1.pie(sizes, explode=explode, labels=labels, autopct='%1.1f%%',
        shadow=False, startangle=90)
ax1.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.

plt.show()
```

```python
fig, ax = plt.subplots(figsize = (10,10))
size = 0.4

vals = np.array([[12,45], [45,23], [66,35]])
vals_sum = vals.sum(axis = 1)
vals_flat = vals.flatten()
cmap = plt.get_cmap('tab20c')
outer_color = cmap(np.arange(8))
inner_color = cmap(np.arange(9,18))

ax.pie(vals_sum, radius = 1.3, colors = outer_color, wedgeprops = 
      dict(width = size, edgecolor = 'w'))

ax.pie(vals_flat, radius = 1.3 - size, colors =  inner_color,
      wedgeprops = dict(width = size, edgecolor = 'w'));
```
