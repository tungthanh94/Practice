### Area and Scatter Plot 

```python
df.head()
```

```python
df.plot(kind = 'area')
```

```python
df.plot(kind = 'area', stacked = False, alpha = 0.2)
```

```python
df.plot.scatter(x = 'sepal_width', y = 'petal_width')
```

### Scatter Matrix and Subplots

```python
from pandas.plotting import scatter_matrix
```

```python
scatter_matrix(df, figsize=(8,8), diagonal = 'kde', color = 'r')
plt.show()
```

```python
df.plot(subplots=True)
plt.show()
```

```python
df.plot(subplots=True, sharex = False)
plt.tight_layout()
plt.show()
```

```python
df.plot(subplots=True, sharex = False, layout = (2, 2), figsize = (6, 6))
plt.tight_layout()
plt.show()
```
