```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
plt.rcParams[]
```

```python
df = pd.read_csv('daily-temperature.csv')
```

```python
df.head()
```

```python
df.info()
```

```python
df['Date'].astype(datetime)
```

```python
df['Date'] = pd.to_datetime(df['Date'])
```

```python
df.set_index('Date', inplace = True)
```

```python
df
```

```python
df.plot(color = 'm', figsize = (14,8))
plt.xlabel('Year')
plt.ylabel('Degree')
plt.title('Temp')
```

```python
plt.figure(figsize = (14,8), dpi = 150)
plt.plot(df['Temp'])
```

```python
df.plot(style = '.', figsize = (14,10))
```

```python
ygr = df.groupby(pd.Grouper(freq = 'Y'))
```

```python
ygrkey = ygr.groups.keys()
```

```python
years = pd.DataFrame()
for key in ygrkey:
    years[key] = ygr.get_group(key)['Temp'].values
```

```python
years.columns = indexs = list(range(1981, 1991))
```

```python
years
```

```python
years.plot(subplots = True, figsize = (14, 40))
plt.tight_layout()
```

```python
plt.matshow(years.T, aspect = 'auto')
```
